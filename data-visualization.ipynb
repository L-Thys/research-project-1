```python
def viz_all():
  fig, ((ax0, ax1, ax2), (ax3, ax4, ax5)) = plt.subplots(nrows=2, ncols=3,
                                    figsize=(8, 5), layout='constrained')
  for model_name in ["bert", "roberta", "xlnet"]:
    ax0.plot(["base", "large"], [all_data["AUL cp"][model_name],all_data["AUL cp"][model_name+"_large"]],label=model_name)
    ax1.plot(["base", "large"], [all_data["AUL ss"][model_name],all_data["AUL ss"][model_name+"_large"]])
    ax4.plot(["base", "large"], [all_data["CrowS-Pairs"][model_name],all_data["CrowS-Pairs"][model_name+"_large"]])
    ax3.plot(["base", "large"], [all_data["CAT bias score"][model_name],all_data["CAT bias score"][model_name+"_large"]])
    ax5.plot(["base", "large"], [all_data["LPBS"][model_name],all_data["LPBS"][model_name+"_large"]])
    ax2.plot(["base", "large"], [all_data["SEAT"][model_name],all_data["SEAT"][model_name+"_large"]])


  # Add some text for labels, title and custom x-axis tick labels, etc.
  fig.legend(loc='upper right')
  ax0.set_title("AUL cp")
  ax1.set_title("AUL ss")
  ax4.set_title("CrowS-Pairs")
  ax3.set_title("CAT bias score")
  ax5.set_title("LPBS")
  ax2.set_title("SEAT")

  plt.show()

viz_all()
```
